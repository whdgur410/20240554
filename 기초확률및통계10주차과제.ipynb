```python
import matplotlib.pyplot as plt

# üéØ Í≤∞Ìï©ÌôïÎ•†ÏßàÎüâÌï®Ïàò Îç∞Ïù¥ÌÑ∞ Ï†ïÏùò
joint_pmf_data = {
    (1, 1): 0.1,
    (1, 2): 0.2,
    (1, 3): 0.2,
    (2, 1): 0.3,
    (2, 2): 0.05,
    (2, 3): 0.15
}

# üìå Ï£ºÎ≥ÄÌôïÎ•†Î∞ÄÎèÑÌï®Ïàò Ï¥àÍ∏∞Ìôî
f_X = {}
f_Y = {}

# üìà XÏùò Ï£ºÎ≥ÄÌôïÎ•†Î∞ÄÎèÑÌï®Ïàò Í≥ÑÏÇ∞
for x, y in joint_pmf_data.keys():
    f_X[x] = f_X.get(x, 0) + joint_pmf_data[(x, y)]

# üìâ YÏùò Ï£ºÎ≥ÄÌôïÎ•†Î∞ÄÎèÑÌï®Ïàò Í≥ÑÏÇ∞
for x, y in joint_pmf_data.keys():
    f_Y[y] = f_Y.get(y, 0) + joint_pmf_data[(x, y)]

# üñºÔ∏è Í∑∏ÎûòÌîÑ ÏÑ§Ï†ï
plt.figure(figsize=(12, 5))

# YÏ∂ï Î≤îÏúÑ ÏÑ§Ï†ï
y_min = 0
y_max = 1

# XÏ∂ï Î≤îÏúÑ ÏÑ§Ï†ï
x_min = 0.5
x_max = 3.5
```

```python
# üîµ XÏùò Ï£ºÎ≥ÄÌôïÎ•†ÏßàÎüâÌï®Ïàò Í∑∏ÎûòÌîÑ
plt.subplot(1, 2, 1)
plt.plot(list(f_X.keys()), list(f_X.values()), 'bo')  # ÌååÎûÄ Ï†ê
plt.vlines(list(f_X.keys()), y_min, list(f_X.values()), colors='blue')  # ÏàòÏßÅÏÑ†
plt.title('Marginal PMF of X')
plt.xlabel('X value')
plt.ylabel('Probability')
plt.xticks(range(int(x_min), int(x_max) + 1))
plt.xlim(x_min, x_max)
plt.ylim(y_min, y_max)

# üî¥ YÏùò Ï£ºÎ≥ÄÌôïÎ•†ÏßàÎüâÌï®Ïàò Í∑∏ÎûòÌîÑ
plt.subplot(1, 2, 2)
plt.plot(list(f_Y.keys()), list(f_Y.values()), 'ro')  # Îπ®Í∞Ñ Ï†ê
plt.vlines(list(f_Y.keys()), y_min, list(f_Y.values()), colors='red')  # ÏàòÏßÅÏÑ†
plt.title('Marginal PMF of Y')
plt.xlabel('Y value')
plt.ylabel('Probability')
plt.xticks(range(int(x_min), int(x_max) + 1))
plt.xlim(x_min, x_max)
plt.ylim(y_min, y_max)

# üìê Î†àÏù¥ÏïÑÏõÉ Ï†ïÎ¶¨ Î∞è Í∑∏ÎûòÌîÑ Ï∂úÎ†•
plt.tight_layout()
plt.show()
```

```python
import numpy as np

# üìö Íµ≠Ïñ¥ ÏÑ±Ï†ÅÍ≥º ÏàòÌïô ÏÑ±Ï†Å Îç∞Ïù¥ÌÑ∞
korean_scores = [75, 65, 50, 70, 80]
math_scores = [65, 70, 70, 55, 60]

# üìä Îç∞Ïù¥ÌÑ∞Î•º NumPy Î∞∞Ïó¥Î°ú Î≥ÄÌôò
data = np.array([korean_scores, math_scores])

# üìà Í≥µÎ∂ÑÏÇ∞ Í≥ÑÏÇ∞ (bias=True: ÌëúÎ≥∏Ïù¥ ÏïÑÎãå Ï†ÑÏ≤¥ Îç∞Ïù¥ÌÑ∞Î°ú Í∞ÑÏ£º)
covariance = np.cov(data, bias=True)[0, 1]

# üñ®Ô∏è Í≤∞Í≥º Ï∂úÎ†•
print("Covariance:", covariance)
```
